```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Reshape

# Function to generate training data for matrix transpose
def generate_data(num_samples, matrix_size):
    X = np.random.rand(num_samples, matrix_size, matrix_size)
    Y = np.transpose(X, axes=(0, 2, 1))  # Transpose each matrix
    return X, Y

# Parameters
num_samples = 1000  # Number of samples to generate
matrix_size = 3     # Size of the matrix (e.g., 3x3)

# Generate data
X_train, Y_train = generate_data(num_samples, matrix_size)

# Neural network model
model = Sequential([
    Flatten(input_shape=(matrix_size, matrix_size)),  # Flatten the input matrix
    Dense(64, activation='relu'),  # Hidden layer with 64 neurons
    Dense(matrix_size * matrix_size),  # Output layer
    Reshape((matrix_size, matrix_size))  # Reshape output to the original matrix shape
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X_train, Y_train, epochs=100, batch_size=32, verbose=1)

# Test the model
X_test, Y_test = generate_data(10, matrix_size)  # Generate some test data
predictions = model.predict(X_test)

# Display results
for i in range(5):
    print(f"Original Matrix:\n{X_test[i]}")
    print(f"True Transpose:\n{Y_test[i]}")
    print(f"Predicted Transpose:\n{predictions[i]}")
    print("-" * 30)
```
